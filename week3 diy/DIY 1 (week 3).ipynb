```python
price_mean = cars.price.mean()
price_std = cars.price.std()
cars['isOutlierPrice'] = np.where(((cars['price']>(price_mean+(price_std*2)))|(cars['price']<(price_mean-(price_std*2)))), 1, 0)
```

```python
mileage_mean = cars.mileage.mean()
mileage_std = cars.mileage.std()
cars['isOutlierMileage'] = np.where(((cars['mileage']>(mileage_mean+(mileage_std*2)))|(cars['mileage']<(mileage_mean-(mileage_std*2)))), 1, 0)
```

```python
color = np.where(cars.isOutlierMileage==1,'red','green')
plt.scatter(cars.price,cars.mileage,color =color, alpha = 0.5)
```
